```python
import pandas as pd
import pymongo
```

```python
client = pymongo.MongoClient("mongo")
database = client['exercices']
collection = database['youtube']
```

```python
df_youtube = pd.read_csv("./data/USvideos.csv")
df_youtube.head(2)
```

## Question 0

```python
import json 
df_category = pd.read_json("./data/US_category_id.json")
a = df_category["items"]
s = pd.DataFrame()
for i in range(len(a)):
    x = json.dumps(a[i])
    df = pd.read_json(x)
    s = pd.concat([s,df]) 
id = s['id']['title']
tittle = s['snippet']['title']
d1 = pd.DataFrame({'category_id': id, 'Title': tittle})
d1
```

```python
df_youtube.head()
```

```python
df = d1.merge(df_youtube)
df.head(15)
```

## Question 5

```python
cur = collection.aggregate([
    {"$group" : {"_id" : "$category_id", "nb_views_cat" : {"$avg" : "$views"}}}
])
list(cur)
```

## Question 6 

```python
cur = collection.aggregate([
    {"$group" : {"_id" : "$channel_title", "nb_likes_channel" : {"$avg" : "$likes"}}},
    {"$sort" : { "nb_likes_channel" : -1}}
    ])
list(cur)
```
